```python
theta_A, theta_B, theta_C = 0.3,0.5,0.8 #Set the success rate for three slot machine differently.
alpha, beta, n= 1, 1, 10 #Set the initial value of alpha, beta and sample size
prior = stats.beta(a=alpha, b=beta)
```

```python
x_A=0
np.random.seed(400) #Set seed for the random variable so that we get the same result of each round of the simulation. 
for i in range(n):
    if stats.binom(n=1, p=theta_A).rvs(size=1) == 1:
        x_A += 1
# Updating the parameters according to simulated results given by binomial(n, p)
alpha_A = alpha+ x_A + alpha - 1
beta_A = beta+n - x_A + beta - 1
print(alpha_A, beta_A)
posterior_A = stats.beta(a=alpha_A,b=beta_A)
posterior_A.pdf(0.5)
```

```python
x_B=0
np.random.seed(400) #Set seed for the random variable so that we get the same result of each round of the simulation.
for i in range(n):
    if stats.binom(n=1, p=theta_B).rvs(size=1) == 1:
        x_B += 1

# Updating the parameters according to simulated results given by binomial(n, p)
alpha_B = alpha+ x_B+ alpha - 1
beta_B = beta+n - x_B + beta - 1
print(alpha_B, beta_B)
posterior_B = stats.beta(a=alpha_B,b=beta_B)
posterior_B.pdf(0.5)
```

```python
x_C=0
np.random.seed(400) #Set seed for the random variable so that we get the same result of each round of the simulation.
for i in range(n):
    if stats.binom(n=1, p=theta_C).rvs(size=1) == 1:
        x_C += 1

# Updating the parameters according to simulated results given by binomial(n, p)
alpha_C = alpha+ x_C + alpha - 1
beta_C = beta+n - x_C + beta - 1
print(alpha_C, beta_C)
posterior_C = stats.beta(a=alpha_C,b=beta_C)
posterior_C.pdf(0.5)
```

```python
#Construct visualization 
import matplotlib.pyplot as plt  
parameter_domain = np.linspace(0,1,1001)
plt.figure(figsize=(12,4)) 
plt.plot(parameter_domain, posterior_A.pdf(parameter_domain), label="machine A")
plt.plot(parameter_domain, posterior_B.pdf(parameter_domain), label="machine B")
plt.plot(parameter_domain, posterior_C.pdf(parameter_domain), label="machine C")
plt.legend()
plt.title("PDF for machines")
plt.xlabel('Success Rate')
plt.ylabel('Probability Density')
#Based on the visualization, machine C has the highest success rate with the simulation, that was been made under unknow hypothesis. 
#This result matched with our original design. 
```
